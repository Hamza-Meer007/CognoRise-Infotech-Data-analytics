### Exploratory Data Analysis

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.countplot(x=df['quality'],data=df)
plt.title('Wine Quality',fontdict={'fontsize':25,'color':'red'})
plt.show()
```

##### Wine has quality of 5,6 in majority

```python
sns.distplot(df['alcohol'])
plt.title('Alcohol Distribution',fontdict={'fontsize':25,'color':'red'})
plt.show()
```

##### The Alcohal has a positively skewed distribution

```python
sns.heatmap(df.corr(),annot=True,fmt='.2f')
plt.title('Correlation Matrix',fontdict={'fontsize':25,'color':'red'})
plt.show()
```

##### 1. Quality has high correlation with Alcohol
##### 2. pH has weak correlation with alcohol
##### 3. Citric acid is highly correlated with fixed acidity

```python
sns.distplot(df['pH'])
plt.title('pH Distribution',fontdict={'fontsize':25,'color':'red'})
plt.show()
```

```python
sns.distplot(df['fixed acidity'])
plt.title('Fixed Acidity Distribution',fontdict={'fontsize':25,'color':'red'})
plt.show()
```

```python
sns.distplot(df['citric acid'])
plt.title('Citric Acid Distribution',fontdict={'fontsize':25,'color':'red'})
plt.show()
```

```python
sns.boxplot(x=df['quality'],y=df['alcohol'],data=df,showfliers=False)
plt.title('Quality vs Alcohol',fontdict={'fontsize':25,'color':'red'})
plt.show()
```

##### Higher the concentration of alcohol means higher the quality of wine.

```python
sns.regplot(x=df['citric acid'],y=df['fixed acidity'],data=df)
plt.title('Citric Acid vs Fixed Acidity',fontdict={'fontsize':25,'color':'red'})
plt.show()
```

##### Citric acid is highly correlated with fixed Acidity

```python
sns.regplot(data=df,x=df['alcohol'],y=df['pH'])
plt.title('Alcohol vs pH',fontdict={'fontsize':25,'color':'red'})
plt.show()
```

##### From the above graph, it is clear that pH and Alcohol are weakly correlated
